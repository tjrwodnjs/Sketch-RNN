```python
import matplotlib.pyplot as plt

def plot_strokes(strokes):
    x, y = 0, 0
    for dx, dy, t in strokes:
        dy *= -1 # correct y direction for ploting
        if t == 0:
            plt.plot([x, x+dx], [y, y+dy], color="gray")
        x, y = x + dx, y + dy
    
    plt.show()
    return
```

```python
for i in range(5):
    plot_strokes(data['train'][i])
```
